```python
# demo load and process the full menu

#make menu and empty list

menu =[]

#read in our .csv with 'with' statement
with open('../data/raw/beanscene_menu.csv',newline="",encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        #print(row)
        menu.append({
            "name":row["name"],
            "price": float(row["price"]),
            "units_sold": int(row["units_sold"]),
            "category":row["category"]
        })
```

```python
for item in menu:
    revenue = item["price"] * item["units_sold"]
    print(f"{item['name']:<14}: ${revenue:.2f}")
```

```python
# with statement to write to a .csv file
with open("../data/processed/menu_results.csv",mode='w',newline="",encoding="utf-8")as f:
        writer = csv.DictWriter(f, fieldnames=['name','revenue','tier'])
        writer.writeheader()
        writer.writerows(results)

print("Yay, it saved!")
```

```python
with open("../data/raw/beanscene_config.json", encoding="utf-8") as f:
    config = json.load(f)

#config
high_threshold = config["performance_thresholds"]["high"]
low_threshold = config["low_performer_threshold"]

#create an empty list to store our menu
menu = []

with open("../data/raw/beanscene_menu.csv",newline="", encoding="utf-8") as f:
    for now in csv.DictReader(f):
        menu.append({
            "name":row["name"],
            "price":float(row['price']),
            "units_sold": int(row["units_sold"]),
            })
    
```

```python
menu
```

```python
# Now our "analysis" on our menu item revenue tiers
for item in menu:
    # print(item)
    # calculate revenue
    revenue = item["price"] * item["units_sold"]
    if revenue >= high_threshold:
        tier = "High"
    elif revenue >= low_threshold:
        tier = "Medium"
    else:
        tier = "Low"
    print(f"{item['name']}: ${revenue:.2f} [{tier}]")
```

```python
summary = {
    "cafe_name": "BeanScene",
    "week": "2024-W04",
    "total_revenue": 1521.75,
    "best_seller": "Latte",
    "items_analysed": 8,
}
```

```python
with open("../data/processed/weekly_summary.json", mode="w", encoding='utf-8') as f:
    json.dump(summary,f, indent=2)
```
